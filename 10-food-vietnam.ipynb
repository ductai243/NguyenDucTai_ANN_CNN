```python
# Preprocessing
train = ImageDataGenerator(rescale=1/255)
image_generator = ImageDataGenerator(rescale=1/255, validation_split=0.2)    

train_dataset = image_generator.flow_from_directory(batch_size=32,
                                                 directory="../input/fooddatasets/10foods/",
                                                 shuffle=True,
                                                 target_size=(150, 150), 
                                                 subset="training",
                                                 class_mode='categorical')

validation_dataset = image_generator.flow_from_directory(batch_size=32,
                                                 directory="../input/fooddatasets/10foods/",
                                                 shuffle=True,
                                                 target_size=(150, 150), 
                                                 subset="validation",
                                                 class_mode='categorical')
```

```python
# Reshape data
import glob
banhcuon = glob.glob('../input/fooddatasets/10foods/Banh cuon/*.*')
banhgio = glob.glob('../input/fooddatasets/10foods/Banh gio/*.*')
banhmi = glob.glob('../input/fooddatasets/10foods/Banh mi/*.*')
banhpia = glob.glob('../input/fooddatasets/10foods/Banh pia/*.*')
bunbohue = glob.glob('../input/fooddatasets/10foods/Bun ho Hue/*.*')
bunrieu = glob.glob('../input/fooddatasets/10foods/Bun rieu/*.*')
cakhoto = glob.glob('../input/fooddatasets/10foods/Ca kho to/*.*')
chaolong = glob.glob('../input/fooddatasets/10foods/Chao long/*.*')
comtam = glob.glob('../input/fooddatasets/10foods/Com tam/*.*')
nemchua = glob.glob('../input/fooddatasets/10foods/Nem chua/*.*')



data = []
labels = []

for i in banhcuon:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(0)
for i in banhgio:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(1)
for i in banhmi:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(2)
for i in banhpia:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(3)

for i in bunbohue:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(4)
for i in bunrieu:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(5)
for i in cakhoto:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(6)
for i in chaolong:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(7)
for i in comtam:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(8)
for i in nemchua:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(9)

data = np.array(data)
labels = np.array(labels)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, test_size=0.2,
                                                random_state=42)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
X_train = X_train.reshape((X_train.shape[0],150,150,3)).astype('float32')/255
X_test = X_test.reshape((X_test.shape[0],150,150,3)).astype('float32')/255

Y_train = to_categorical(Y_train,10)
Y_test = to_categorical(Y_test,10)
```

```python
# Create model
from keras.layers import Conv2D, MaxPooling2D
model = Sequential()
model.add(Conv2D(32,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same',input_shape=(150,150,3)))
model.add(Conv2D(32,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 64 lan tich chap
model.add(Conv2D(64,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 128 lan tich chap
model.add(Conv2D(128,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(256,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 256 lan tich chap
model.add(Conv2D(256,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))


from keras.layers import Dense, Activation, Flatten
model.add(Flatten())
model.add(Dense(128, activation = 'relu', kernel_initializer='he_uniform'))
model.add(Dense(10))
model.summary()
```

```python
# Compile model
model.compile(
    optimizer='rmsprop',
    loss='mse',
    metrics=["accuracy"])
# Training
history = model.fit(X_train,Y_train, epochs=50, batch_size=128,verbose=1, validation_split=0.3)


# model.compile(loss='mse',optimizer=RMSprop(),metrics=['accuracy'])
# history = model.fit(X_train, Y_train, epochs=50, batch_size =128,validation_data=(X_test,Y_test) , verbose = 1)
```
